# 1. Scrapping data from Wikipedia

```python
#import required Libraries

import pandas as pd # library for data analsysis

print('Libraries imported.')
```

```python
#scrapping table from wikipedia

url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

df=pd.read_html(url, header=0)[0]

df.head(12)
```

```python
#drop rows which has 'Not assigned' value in column 'Borough'

df = df.drop(df[df['Borough']=='Not assigned'].index)
df
```

```python
#check how many rows there are with 'Not assigned' value under 'Neighbourhood' column

df['Neighbourhood'].value_counts()
```

```python
#Replace 'Not assigned' value in 'Neighbourhood' column to its corresponding 'Borough' value

df = df.replace({'Neighbourhood': {'Not assigned': 'Queen\'s Park'}})
df
```

```python
#check if replacement is correct

df['Neighbourhood'].value_counts()
```

```python
#view dataset

df.head()
```

```python
#Merge duplicate Postcodes

df = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join)
```

```python
#Convert df to DataFrame and rest index

df = df.to_frame().reset_index()
```

```python
#Analyzing each Borough

# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Borough'] = toronto_venues['Borough'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

```python
#shape of dataset

toronto_onehot.shape
```

```python
# group the rows by boroughs and the frequency of their occurrence

toronto_grouped = toronto_onehot.groupby('Borough').mean().reset_index()
toronto_grouped
```

```python
#Print each Borough with top 3 venues

num_top_venues = 3

for hood in toronto_grouped['Borough']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Borough'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
#Putting it into a dataframe

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
#Popular

num_top_venues = 5

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Borough']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
borough_venues_sorted = pd.DataFrame(columns=columns)
borough_venues_sorted['Borough'] = toronto_grouped['Borough']

for ind in np.arange(toronto_grouped.shape[0]):
    borough_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

borough_venues_sorted.head()
```
